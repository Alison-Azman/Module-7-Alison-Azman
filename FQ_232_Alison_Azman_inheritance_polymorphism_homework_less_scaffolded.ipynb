## Problem 1: Build an Animal Hierarchy

Create a parent class named `Animal`.

### The `Animal` class must:
- store a name
- store an age
- include a method named `speak()` that returns a general animal sound
- include a method named `move()` that returns a general description of movement
- include a method named `describe()` that returns a short sentence about the animal

Then create these child classes:

- `Dog`
- `Cat`
- `Rabbit`

### Child class requirements
Each child class must:
- inherit from `Animal`
- use the parent constructor appropriately
- override `speak()`
- override `move()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `speak()`
- each object gives a different result for `move()`
- `describe()` works for each object


```python
# Write your code for Problem 1 here.
class Animal:

  def __init__(self, name:str, age:int):
    self.name=name
    self.age=age

  def speak(self):
    return f"Animal sound"

  def move(self):
    return f"Walk walk"

  def describe(self):
    return f"{self.name} is {self.age} years old"

class Dog(Animal):
  def __init__(self, name, age):
    super().__init__("dog",age)
    self.name=name
    self.age=age

  def speak(self):
    return f"Woof Woof"

  def move(self):
    return f"Run run"


class Cat(Animal):
  def __init__(self, name, age):
    super().__init__("cat",age)
    self.name=name
    self.age=age

  def speak(self):
    return f"Meow Meow"

  def move(self):
    return f"Tiptoe"

class Rabbit(Animal):

  def __init__(self, name, age):
    super().__init__("rabbit",age)
    self.name=name
    self.age=age

  def speak(self):
    return f"Crunch crunch"

  def move(self):
    return f"Leap leap"

dog= Dog("Lucky", 4)
cat=Cat("Pecry",23)
rabbit=Rabbit("Bob", 2)

# I wrote this before I saw later in the questions that it is part of a different question
animals=[dog,cat,rabbit]
for animal in animals:
  print(animal.describe())
  print(animal.speak())
  print(animal.move())
  print()

#Since it looks like we only were supposed to do the loop later I wrote out all the code
# from the loop
print(dog.describe())
print(dog.speak())
print(dog.move())

print()
print(cat.describe())
print(cat.speak())
print(cat.move())

print()
print(rabbit.describe())
print(rabbit.speak())
print(rabbit.move())
```

## Problem 2: Build a Vehicle Hierarchy

Create a parent class named `Vehicle`.

### The `Vehicle` class must:
- store a brand
- store a speed
- include a method named `move()` that returns a general statement about transportation
- include a method named `sound_horn()` that returns a general horn sound
- include a method named `describe()` that returns a short sentence about the vehicle

Then create these child classes:

- `Car`
- `Truck`
- `Motorcycle`

### Child class requirements
Each child class must:
- inherit from `Vehicle`
- use the parent constructor appropriately
- override `move()`
- override `sound_horn()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `move()`
- each object gives a different result for `sound_horn()`
- `describe()` works for each object


```python
# Write your code for Problem 2 here.
class Vehicle:

  def __init__(self, brand:str, speed:int):
    self.brand=brand
    self.speed=speed

  def move(self):
    return f"The vehicle is moving"

  def sound_horn(self):
    return f"Honk Honk"

  def describe(self):
    return f"The {self.brand} is going {self.speed}mph."

class Car(Vehicle):

  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return f"The car is going Vrooom"

  def sound_horn(self):
    return f"Beep Beep"

class Truck(Vehicle):

  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return f"The truck goes Vroom"

  def sound_horn(self):
    return f"HOOONNK"

class Motorcycle(Vehicle):

  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return f"The motorcycle goes Vrooooooooooom"

  def sound_horn(self):
    return f"Meep Meep"

car= Car("Toyota", 45)
truck=Truck("Saab",30)
motorcycle=Motorcycle("Kawasaki", 75)

#same thing here, I had already written the loop before realizing it is part of a different question
vehicles=[car, truck, motorcycle]
for vehicle in vehicles:
  print(vehicle.describe())
  print(vehicle.move())
  print(vehicle.sound_horn())
  print()
```
