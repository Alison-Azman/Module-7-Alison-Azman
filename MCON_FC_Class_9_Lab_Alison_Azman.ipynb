Inheritance Polymorphism

Polymorphism is the ability to define a method
in a superclass and then define a method with
the same name in a subclass. When a subclass
defines a method with the same name as one in
the superclass, the subclass's method
overrides the superclass's method.

Polymorphism also means that the program
automatically calls the correct version of the
method based on the type of object being used.
If the object is created from a subclass, the
subclass's version of the method will run. If the
object is created from the superclass, the
superclass's version of the method will run.

Example : honk()

Notice this line of code :

def show animal info (creature: Animal) -> None:

We use Animal because it is the parent class of all the animal subclasses. That means creature
can be a Dog, Cat, or any other subclass of Animal, and the function will still work.

If we wrote creature: Dog instead, then passing a Cat would not make sense - a Cat is not a
Dog.

So when your code needs to work with many different subclasses, you should type it using the
most general superclass that still makes sense. This keeps your code flexible and supports
polymorphism.

In Python, type hints don't stop the code from running - but they do express your intent and help tools
catch mistakes. For polymorphism, we hint the parent type so the function is clearly meant to accept
any subclass.
