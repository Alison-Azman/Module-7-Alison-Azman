MRO

Method Resolution Order

Instance  method- self, one object

Class method- cls, the class

Static method- give you a response of some kind, even though no access to self or cls it is part of class because considered part of what you are doing

```python
class Student:
  school_name= "Touro" #class method attribute

  def __init__(self, name):
    self.name=name


  def get_name(self):
    return f"{self.name} is a student."


  @classmethod
  def get_school_name(cls):
    return f"The student attends {cls.school_name}."

  #static method
  @staticmethod
  def is_valid_gpa(gpa:float):
    return 0.0 <gpa <= 4.3

#Don't need to instanciate it
print(Student.is_valid_gpa(5.0))
student1= Student("Alison")

print(student1.get_name())
print(student1.get_school_name())
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
  def __init__(self, name:str):
    self.name=name

  def get_name(self):
    return self.name

  @abstractmethod
  def speak(self):
    pass


class Cat(Animal):
  def __int__ (self, name):
    self.name=name
    super().__init__(name)

  def speak(self):
    return 'meow'


class Dog(Animal):

  @abstractmethod
  def speak(self):
    pass
class Pyr(Dog):
  def speak(self):
    return "woof"

my_animal=Cat("Frizzy")
my_animal.speak()
#Now it works becuase speak is created in the child class
#my_dog=Dog() this won't wprk because does not have the method speak
my_dog=Pyr("Lucky")

```

```python
#wordgame abstract class

from abc import ABC, abstractmethod

class WordGameController(ABC):
  @abstractmethod
  def play(self):
    pass

  @abstractmethod
  def make_guess(self, letter:str):
    pass

  @abstractmethod
  def check_game_over(self)->bool:
    pass


class TinyWordGame(WordGameController):
  def __init__ (self, secret_word:str):
    self.secret_word=secret_word
    self.guesses=set()

  def play(self):
    print("Welcome to the Tiny Word Game, Guess the Secret Letter")

  def make_guess(self, letter: str):
    self.guesses.add(letter)
    return letter in self.secret_word


  def check_game_over(self) -> bool:
    return all(letter in self.guesses for letter in self.secret_word)

game = TinyWordGame("flower")
game.play()
game.make_guess("c")
print(game.check_game_over())
```

```python
class HintWordGame(WordGameController): #accept a hint
  def __init__ (self, secret_word:str, hint:str):
    self.secret_word=secret_word
    self.guesses=set()
    self.hint=hint

  def play(self):
    print("Welcome to the Tiny Word Game, Guess the Secret Letter")
    print(f"Your special hint is that the word contains the letter {self.hint}")

  def make_guess(self, letter: str):
    self.guesses.add(letter)
    return letter in self.secret_word


  def check_game_over(self) -> bool:
    return all(letter in self.guesses for letter in self.secret_word)

game = HintWordGame("flower", "r")
game.play()
game.make_guess("c")
print(game.check_game_over())
```

```python
class LimitedGuessWordGame(WordGameController):

   def __init__(self, secret_word: str, max_guesses: int):
        self.secret_word = secret_word
        self.max_guesses = max_guesses
        self.guesses = set()
        self.guess_count = 0
   def play(self):
    print("Welcome to the Tiny Word Game, Guess the Secret Letter")
    print(f"You have {self.max_guesses} guesses for your Game")


   def make_guess(self, letter: str):
        self.guesses.add(letter)
        self.guess_count += 1
        self.how_many_guesses_left()
        return letter in self.secret_word

   def how_many_guesses_left(self):
    print(f"You have used {self.guess_count} guess(es) out of {self.max_guesses}")

   def check_game_over(self):
        guesses_left = self.max_guesses - self.guess_count
        won = all(letter in self.guesses for letter in self.secret_word)
        return guesses_left, won

game = LimitedGuessWordGame("flower", 5)
game.play()
game.make_guess("c")
print(game.check_game_over())
```

```python
class Flyer:
  def __init__(self, animal_name:str):
    self.animal_name=animal_name

  @staticmethod
  def fly():
    print("I am flying")


class Swimmer:
  def __init__(self, swimmer_name:str):
    self.swimmer_name = swimmer_name

  @staticmethod
  def swim():
    print("I am swimming")


class FlySwimmer(Flyer, Swimmer):

  def __init__(self, flyer_name:str, swimmer_name:str):
    Flyer.__init__(self, flyer_name)
    Swimmer.__init__(self, swimmer_name)

duck=FlySwimmer("Daffy", "Swimming")
duck.swim()
duck.fly()
```

```python
class Vehicle:

  def __init__(self, brand: str, speed: int):
    self.brand=brand
    self.speed=speed

  @staticmethod
  def drive():
    print(f"I am driving")


class Flying(Vehicle):

  def __init__(self, alt:int, airport_dest:str):
    self.alt=alt
    self.airport_dest = airport_dest

  @staticmethod
  def flying():
    print("I am flying")


class FlyingCar(Flying, Vehicle):
    def __init__(self, brand: str, speed: int, alt: int, airport_dest: str):
      self.brand=brand
      self.speed=speed
      self.alt=alt
      self.airport_dest = airport_dest

car= FlyingCar("Mazda", 45, 100, "MIA")
car.drive()
car.flying()

```
