## Question 1: Inheritance with Employees

Create a parent class named `Employee`.

### `Employee` must:
- store an employee name
- store a department
- include a method named `work()` that returns a general work message

Then create a child class named `Manager`.

### `Manager` must:
- inherit from `Employee`
- use the parent constructor
- override `work()` so that it returns a manager-specific message

### After writing your classes
Create a `Manager` object and demonstrate that:
- the object stores its data correctly
- the overridden `work()` method is being used


````python
# Write your answer for Question 1 here.
class Employee:
  def __init__(self, name, dept):
    self.name=name
    self.dept=dept

  def work(self):
    return f"{self.name}, don't forget to do your job"

class Manager(Employee):
  def __init__(self, name, dept):
    super().__init__(name, dept)

  def work(self):
    return f"{self.name}, make sure you oversee everyone"

employee=Employee("Shani", "ABA")
print(employee.work())
manager=Manager("Alison", "Insurance")
print(manager.work())
````


## Question 2: Choose What Belongs in the Parent

Below is a group of possible attributes and behaviors for electronic devices:

- brand
- battery_percent
- charge()
- take_photo()
- make_call()

Create a parent class named `Device` using only the information that should reasonably be shared by many devices.

Then create a child class named `Phone`.

### Requirements
- `Device` should store shared data and include at least one shared behavior
- `Phone` should inherit from `Device`
- `Phone` should add or override behavior that makes sense for a phone

### After writing your classes
Create a `Phone` object and demonstrate that inherited behavior works correctly.


````python
# Write your answer for Question 2 here.
class Device:
  def __init__(self,brand, battery_percent):
    self.brand=brand
    self.battery_percent=battery_percent

  def charge(self):
    self.battery_percent=100
    return f"Your {self.brand} device's battery is {self.battery_percent}%"

  def take_photo(self):
    return f"Picture was taken"


class Phone(Device):

  def __init__(self,brand, battery_percent):
    super().__init__(brand, battery_percent)

  def take_photo(self):
    return f"Snap/Click, picture was taken"

  def make_call(self):
    return f"Ring ring..."

phone=Phone("Apple", 86)
print(phone.battery_percent)
print(phone.charge())
print(phone.take_photo())
print(phone.make_call())



````


## Question 3: Polymorphism with School Roles

Create a parent class named `SchoolMember` with a method named `introduce()`.

Then create these child classes:
- `Teacher`
- `Student`

Each child class should override `introduce()` in its own way.

### Then:
1. Create one `Teacher` object and one `Student` object
2. Store them together in a list
3. Use one loop to call `introduce()` on each object

## Goal
Your output should clearly demonstrate polymorphism.


````python
# Write your answer for Question 3 here.
class SchoolMember:
  def __init__(self, name):
    self.name=name

  def introduce(self):
    return f"Hello, my name is {self.name}."

class Teacher(SchoolMember):
  def __init__(self,name):
    super().__init__(name)

  def introduce(self):
    return f"Hello, I am {self.name} and I am a teacher."


class Student(SchoolMember):
  def __init__(self,name):
    super().__init__(name)

  def introduce(self):
    return f"Hello, I am {self.name} and I am a student."

teacher=Teacher("Dr. Smith")
student=Student("Alison")

school_members=[teacher,student]

for member in school_members:
  print(member.introduce())


#Or without having name:
# Write your answer for Question 3 here.
class SchoolMember:
  def __init__(self):
    pass
  def introduce(self):
    return f"Hello, I am part of the school."

class Teacher(SchoolMember):
  def __init__(self):
    super().__init__()

  def introduce(self):
    return f"Hello, I am a teacher."


class Student(SchoolMember):
  def __init__(self):
    super().__init__()

  def introduce(self):
    return f"Hello, I am a student."

teacher=Teacher()
student=Student()

school_members=[teacher,student]

for member in school_members:
  print(member.introduce())
````
